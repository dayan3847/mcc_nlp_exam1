```python
from typing import List
from pandas import DataFrame
import pandas as pd
import spacy
import openpyxl
```

```python
file: str = "corpus_primer_parcial.xlsx"
df: DataFrame = pd.read_excel(file)
my_texts: List[str] = df.iloc[:, 0].tolist()
my_texts_count: int = len(my_texts)
my_texts
```

```python
# Ejercicio 1
# Pasar a minúsculas todo el texto

def to_lower(text: str) -> str:
    return text.lower()


my_texts_lower = [to_lower(text) for text in my_texts]
my_texts_lower
```

```python
# Ejercicio 2
# Quitar los nombres propios que aparezcan y sustituirlos por una etiqueta llamada #Persona

nlp = spacy.load("es_core_news_sm")


def get_names(text: str) -> set[str]:
    result: set[str] = set()
    doc = nlp(text)
    for ent in doc.ents:
        if 'PER' == ent.label_:
            result.add(ent.text)
    return result


detected_names: List[set[str]] = []
for text in my_texts_lower:
    detected_names.append(get_names(text))
print(detected_names)


def remove_names(text: str) -> str:
    doc = nlp(text)
    for ent in doc.ents:
        if 'PER' == ent.label_:
            detected_names.append(ent.text)
            text = text.replace(ent.text, '#Persona')
    return text


my_texts_lower_no_names = [remove_names(text) for text in my_texts_lower]
my_texts_lower_no_names
```
